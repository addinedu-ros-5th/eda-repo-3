```python
import folium

def folium_gasan_station(df1,m):
    
    circle_options = {
    10000: {'color': 'red', 'fill_color': 'red', 'fill_opacity': 0.2, 'popup': '10km radius'},
    15000: {'color': 'orange', 'fill_opacity': 0.5, 'popup': '15km radius'},
    20000: {'color': 'yellow', 'fill_opacity': 0.5, 'popup': '20km radius'},
    25000: {'color': 'green', 'fill_opacity': 0.5, 'popup': '25km radius'},
    30000: {'color': 'blue', 'fill_opacity': 0.5, 'popup': '30km radius'}}

    df_standard = df1[df1['역명'] == '가산디지털단지']
    df_standard.reset_index(drop=True, inplace=True)
    folium.Marker([df_standard['위도'][1],df_standard['경도'][1]], 
                icon =folium.Icon(icon= 'star', prefix='fa',color=df2['색상'][6])
                ).add_to(m)
    folium.Marker([df_standard['위도'][0],df_standard['경도'][0]], 
                icon =folium.Icon(icon= 'star', prefix='fa',color=df2['색상'][0])
                ).add_to(m)
    
    # 반복문을 사용하여 각 반경의 원을 추가
    for radius, options in circle_options.items():
        folium.Circle(
            location=[df_standard['위도'][1], df_standard['경도'][1]],
            radius=radius,
            **options
        ).add_to(m)
    
def folium_cityhall_station(df1,m):
    
    circle_options = {
    10000: {'color': 'red', 'fill_color': 'red', 'fill_opacity': 0.2, 'popup': '10km radius'},
    15000: {'color': 'orange', 'fill_opacity': 0.5, 'popup': '15km radius'},
    20000: {'color': 'yellow', 'fill_opacity': 0.5, 'popup': '20km radius'},
    25000: {'color': 'green', 'fill_opacity': 0.5, 'popup': '25km radius'},
    30000: {'color': 'blue', 'fill_opacity': 0.5, 'popup': '30km radius'},
    35000: {'color': 'darkblue', 'fill_opacity': 0.5, 'popup': '35km radius'},
    40000: {'color': 'purple', 'fill_opacity': 0.5, 'popup': '40km radius'},
    45000: {'color': 'gray', 'fill_opacity': 0.5, 'popup': '45km radius'},
    50000: {'color': 'white', 'fill_opacity': 0.5, 'popup': '50km radius'},
    58000: {'color': 'pink', 'fill_opacity': 0.5, 'popup': '58km radius'},
    66000: {'color': 'black', 'fill_opacity': 0.5, 'popup': '66km radius'}}

    df_standard = df1[df1['역명'] == '시청']
    df_standard.reset_index(drop=True, inplace=True)
    folium.Marker([df_standard['위도'][1],df_standard['경도'][1]], 
                icon =folium.Icon(icon= 'star', prefix='fa',color=df2['색상'][1])
                ).add_to(m)
    folium.Marker([df_standard['위도'][0],df_standard['경도'][0]], 
                icon =folium.Icon(icon= 'star', prefix='fa',color=df2['색상'][0])
                ).add_to(m)
    
    # 반복문을 사용하여 각 반경의 원을 추가
    for radius, options in circle_options.items():
        folium.Circle(
            location=[df_standard['위도'][0], df_standard['경도'][0]],
            radius=radius,
            **options
        ).add_to(m)
```

```python
numlist=[]

for 호선 in sorted(df1['호선'].unique()):
    count = (df1['호선'] == 호선).sum()
    print(f"{호선}은 {count}개")
    numlist.append(count)
```
