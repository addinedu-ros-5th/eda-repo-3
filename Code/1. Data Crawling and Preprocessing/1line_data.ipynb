```python
from selenium import webdriver
import time

current_url=[]
names = df_filtered['역명'][0:3]+'역'
for name in names:
    
   url = f"https://www.google.com/maps/search/{name}"

   driver = webdriver.Chrome()

   driver.get(url)
 
   time.sleep(5)
   current_url.append(driver.current_url)

   print("현재 URL:", current_url)

   driver.quit()
```
