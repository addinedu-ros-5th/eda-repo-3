```python

numlist=[]

for 호선 in sort_list:
    count = (df3_sorted['호선'] == 호선).sum()
    print(count)
    numlist.append(count)
```

```python
test = df3_sorted.set_index('호선')
test['경도']
```

```python
m = folium.Map(location=[37.58, 127.0], zoom_start=10)

icon_colors = ['black', 'darkblue', 'black', 'gray', 'blue', 'darkred', 'lightgreen', 'purple', 'red', 'green', 'lightred', 'darkgreen', 'beige', 'white', 'lightblue', 'orange', 'pink', 'lightgray', 'lightgreen']
icon_counts = numlist[:len(numlist)] 

# 반복문을 통해 아이콘 추가
start_index = 0
for color, count in zip(icon_colors, icon_counts):
    # 해당 색상의 아이콘을 해당 개수만큼 추가
    for i in range(start_index, start_index + count):
        folium.Marker([test['위도'][i], test['경도'][i]], 
                      icon=folium.Icon(color=color),
                      popup=test['역명'][i]
                     ).add_to(m)
    # 다음 색상의 시작 인덱스를 설정
    start_index += count

# 지도 출력
m
```
