```python
import pandas as pd
import numpy as np
from glob import glob
import os
```

```python
path="../data/seoul/"
types=["수단통행량","목적통행량"]
years=['2022','2023','202401','202402']
times = []
su_col_names = ['시도','연월','이용자유형','교통수단','합계 발생량','합계 도착량']    #수단통행량 컬럼
mok_col_names = ['시도','이용자 유형','연월','합계 발생량','합계 도착량']          #목적통행량 컬럼
time_col_names = []
for time in range(4,25):
    time_col_names.append(str(time)+'시 발생량')
    time_col_names.append(str(time)+'시 도착량')
for time in range(1,4):
    time_col_names.append(str(time)+'시 발생량')
    time_col_names.append(str(time)+'시 도착량')
```

```python
seoul_su_data=[]                #수단통행량
for year in years:
        name="이용량 지표(수단통행량)_{}.xlsx".format(year)
        seoul_su_data.append(pd.read_excel(path+name,header=1,usecols="B:BC",names=su_col_names+time_col_names))
```

```python
seoul_su_data[3].head()
```

```python
seoul_mok_data=[]               #목적통행량
for year in years:
        name="이용량 지표(목적통행량)_{}.xlsx".format(year)
        seoul_mok_data.append(pd.read_excel(path+name,header=1,usecols="B:BB",names=mok_col_names+time_col_names))
```

```python
seoul_mok_data[3].head()
```

```python
for su in seoul_su_data:        #수단통행량에서 항목 대신 0으로 된것 치환
    j=0
    sido_temp = 0
    year_temp = 0
    user_type = 0
    for j in range(len(su)):
        if su["시도"][j] != 0:
            sido_temp = su["시도"][j]
        elif su["시도"][j] == 0 and sido_temp != 0:
            su["시도"][j] = sido_temp

        if su["연월"][j] != 0:
            year_temp = su["연월"][j]
        elif su["연월"][j] == 0 and year_temp != 0:
            su["연월"][j] = year_temp
        
        if su["이용자유형"][j] != 0:
            user_type = su["이용자유형"][j]
        elif su["이용자유형"][j] == 0 and user_type != 0:
            su["이용자유형"][j] = user_type
        
        
```
