```python
import trafilatura
import pandas as pd


# 위키백과 페이지 URL
url = "http://www.seoulmetro.co.kr/kr/page.do?menuIdx=354"

# HTML 내용 가져오기
html = trafilatura.fetch_url(url)

# HTML에서 테이블 추출
tables = trafilatura.extract(html, include_tables=True)
tables
```

```python
table_frame = tables.split('(서울광역ㆍ경기좌석/직행ㆍ인천좌석/광역 버스는 30km초과시)')[1]
```

```python
lines = table_frame.strip().split('\n')
data = [line.split('|')[1] for line in lines[0:]]

data
```

```python
filtered_data = [item for item in data if item not in ['선ㆍ후불 교통카드', '1회용 교통카드']]

print(filtered_data)
```

```python
grouped_data = [filtered_data[i:i+6] for i in range(0, len(filtered_data), 6)]

# 데이터프레임으로 변환합니다.
df = pd.DataFrame(grouped_data)
df
```

```python
df.info()
```

```python
df.columns = df.iloc[0]
df = df.drop(0)
df
```

```python
df = df.apply(lambda x: x.str.replace(',', ''))
df
```

```python
df.info()
```

```python
df[['어른', '청소년', '어린이']] = df[['어른', '청소년', '어린이']].astype(int)
```
